```python
def filter_list(some_list: list):
    return [i for i in some_list  if (type(i) == int )]
```

```python
def first_non_repeating_letter(initial_string: str):
    lower_string = initial_string.lower()
    for character in lower_string:
        if lower_string.count(character) == 1: 
            return initial_string[lower_string.index(character)]
    return ""
```

```python
def digital_root(number: int):
    if(number//10):
        summ = 0 
        digits = []
        while number != 0: 
            digits.append(number%10)
            summ += digits[-1]
            number //= 10
        print(f"{' + '.join(list(map(str, digits[::-1])))} \nSum: {summ}")
        number = digital_root(summ)
    else:
        print(f"Result: {number}")
    return number
```

```python
# First output is a number of pair, second - pairs 
def pair_with_target_sum(some_list: list, target: int):
    pairs = [(some_list[i], some_list[j]) for i in range(len(some_list)) 
                                         for j in range(i+1, len(some_list)) 
                                         if some_list[i] + some_list[j] == target]
    return len(pairs), pairs
```

```python
from operator import itemgetter
def sort_and_upper_friend_list(friend_list: str):
    if(friend_list == ""): return ""
    friend_list = friend_list.upper()
    friend_list = friend_list.split(";")
    friend_list = [tuple(pair.split(":")[::-1]) for pair in friend_list]
    friend_list = sorted(friend_list, key= itemgetter(0,1))
    return friend_list
```

```python
def rearranged(digits: str):
    digits_bigger_than_first = [digit for digit in digits if digit > digits[0]]
    digits_bigger_than_first.append('99') # if there is no bigger digit than first
    min_digit = min(digits_bigger_than_first)
    if(min_digit == '99'):
        return ""
    new_number = min_digit
    digits = digits.replace(min_digit, '', 1)
    digits = ''.join(sorted(digits))
    new_number += digits
    return new_number

def nextBigger(number: int):
    number = str(number)
    for i in range(2, len(number)+1):
        last_part = rearranged(number[-i:])
        if(last_part != ""):
            return int(number[0:-i] + last_part)
    return -1
    
```

```python
def number_to_IPv4(number):
    number_bin = '{:032b}'.format(number)
    octers_bin = [number_bin[i*8:(i+1)*8] for i in range(4)]
    octers = ".".join([str(int(octer,2)) for octer in octers_bin])
    return octers
```
